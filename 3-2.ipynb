# 데이터 가져오기

```python
not_exercise = pd.read_csv('./data/reason_not_exercise_csv_20230119011728.csv', header=1)
not_exercise.head()
```

# 데이터 가공

```python
# '시점' column 삭제
not_exercise.drop(columns='시점', inplace=True)
not_exercise
```

```python
# 지역대분류, 지역소분류 삭제
not_exercise.drop(index=range(22, 52), inplace=True)
not_exercise
```

```python
not_exercise['구분별(1)'] == '성별'
```

```python
not_ex_sex = not_exercise[not_exercise['구분별(1)']=='성별'].copy()
not_ex_sex
```

---

## DataFrame.copy()

```python
test_a = [1, 2, 3, 4]
test_a
```

```python
test_b = test_a
test_b
```

```python
test_b[2] = 0 
test_b
```

```python
test_a
```

* test_a 도 같이 바뀌었음
* [1, 2, 3, 4] 라는 공간을 test_a, test_b 가 가리키고 있음. 그래서 test_b를 변경하는 것은 test_a를 변경하는 것과 같다.

```python
test_c = test_a.copy()
test_c
```

```python
test_c[2]= 100
test_c
```

```python
test_a
```

* DataFrame.copy()는 DataFrame를 새로 하나 만드는 것이다. 주소가 다름.

```python
not_ex_sex.drop(columns='구분별(1)', inplace=True)
not_ex_sex
```

```python
not_ex_sex.set_index('구분별(2)', inplace=True)
not_ex_sex
```

# 한글 설정

```python
import matplotlib.pyplot as plt
%matplotlib inline

from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False

# 한글 설정
f_path = 'c:/Windows/Fonts/malgun.ttf'
font_name = font_manager.FontProperties(fname=f_path).get_name()
rc('font', family=font_name)
```

# 그래프 그리기

```python
f, ax = plt.subplots(1, 2, figsize=(16, 8))

not_ex_sex['운동을 할 충분한 시간이 없어서'].plot.pie(explode=[0, 0.02], ax=ax[0], autopct='%1.1f%%')
ax[0].set_title('운동을 할 충분한 시간이 없어서')
ax[0].set_ylabel('')

not_ex_sex['함께 운동을 할 사람이 없어서'].plot.pie(explode=[0, 0.02], ax=ax[1], autopct='%1.1f%%')
ax[1].set_title('함께 운동을 할 사람이 없어서')
ax[1].set_ylabel('')

plt.show()
```

```python
not_ex_age = not_exercise[not_exercise['구분별(1)']=='연령별'].copy()
not_ex_age.drop(columns='구분별(1)', inplace=True)
not_ex_age.set_index('구분별(2)', inplace=True)
not_ex_age
```

```python
f, ax = plt.subplots(1, 4, figsize=(16, 8))

explode_setting = [0.02, 0.02, 0.02, 0.02, 0.02, 0.02] # 항목의 갯수만큼

not_ex_age['운동을 할 충분한 시간이 없어서'].plot.pie(explode=explode_setting, 
                                         ax=ax[0], 
                                         autopct='%1.1f%%')
ax[0].set_title('운동을 할 충분한 시간이 없어서')
ax[0].set_ylabel('')


not_ex_age['함께 운동을 할 사람이 없어서'].plot.pie(ax=ax[1], 
                                         autopct='%1.1f%%')
ax[1].set_title('함께 운동을 할 사람이 없어서')
ax[1].set_ylabel('')


not_ex_age['운동을 할 만한 장소가 없어서'].plot.pie(explode=explode_setting, 
                                         ax=ax[2], 
                                         autopct='%1.1f%%')
ax[2].set_title('운동을 할 만한 장소가 없어서')
ax[2].set_ylabel('')

not_ex_age['운동을 싫어해서'].plot.pie(explode=explode_setting, 
                                         ax=ax[3], 
                                         autopct='%1.1f%%')
ax[3].set_title('운동을 싫어해서')
ax[3].set_ylabel('')
plt.show()
```
